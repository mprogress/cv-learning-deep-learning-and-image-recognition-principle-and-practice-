```python
##################4.2.1   Sigmoid 函数###################

###########表达式  p = 1/(1+e^(-x))


import numpy as np
import matplotlib.pyplot as plt

def sigmod(x):
    y = 1.0/(1.0 + np.exp(-x))
    return y

plot_x = np.linspace(-10,10,100)
plot_y = sigmod(plot_x)
plt.plot(plot_x , plot_y)
plt.show()
```

```python
###########  4.2.2 最优化方法 ——梯度下降法####################

################ 损失函数###############

import numpy as np
import matplotlib.pyplot as plt

plot_x = np.linspace(-1,6,141)
plot_y = (plot_x - 2.5)**2 -1           #定义一个二次方程的损失函数
plt.scatter(plot_x[5],plot_y[5],color = 'r')   #设置起始点，红色
plt.plot(plot_x ,plot_y)
```

```python
#通过大量数据训练和调整自变量参数大小，使损失函数的值最小，因此使用求导，在梯度下降中*学习率η（步长）  .
#η是一个介于0-1之间的数值

########### 梯度下降过程 ###############
import numpy as np
import matplotlib.pyplot as plt

def J(theta):   #   损失函数
    return (theta -2.5)**2 -1
    
def dJ(theta):        #损失函数的导数
    return 2*(theta -2.5)

theta = 0.0   #初始点
theta_history = [theta]
eta = 0.1   #步长
epsilon = 1e-8  #精度问题或者eta的设置无法使得导数为0
while True:
    gradient = dJ(theta)    #求导数
    last_theta = theta      #先记录下上一个theta的值
    theta = theta - eta * gradient     #得到一个新的theta
    theta_history.append(theta)
    if(abs(J(theta) - J(last_theta)) < epsilon):
        break

        
plt.plot(plot_x,J(plot_x),color = 'r')
plt.plot(np.array(theta_history),J(np.array(theta_history)),color = 'b',marker = 'x')
plt.show()
print(len(theta_history))     #一共走了多少步
    
```

```python
##################### 4.2.3 学习率（步长）η的分析################
#η是一个介于0-1之间的数值
#从以下实验中我们可以肯出，η是一个需要认真调整的参数。过小会导致收敛过慢，而过大有可能会导致模型不收敛。
```

```python
###1. 当将η设置为0.01时（原来为0.1）  步数为424，整个学习的速度就减慢了，最终找到了最优解
import numpy as np
import matplotlib.pyplot as plt

def J(theta):   #   损失函数
    return (theta -2.5)**2 -1
    
def dJ(theta):        #损失函数的导数
    return 2*(theta -2.5)

theta = 0.0   #初始点
theta_history = [theta]
eta = 0.01   #步长
epsilon = 1e-8  #精度问题或者eta的设置无法使得导数为0
while True:
    gradient = dJ(theta)    #求导数
    last_theta = theta      #先记录下上一个theta的值
    theta = theta - eta * gradient     #得到一个新的theta
    theta_history.append(theta)
    if(abs(J(theta) - J(last_theta)) < epsilon):
        break

        
plt.plot(plot_x,J(plot_x),color = 'r')
plt.plot(np.array(theta_history),J(np.array(theta_history)),color = 'b',marker = 'x')
plt.show()
print(len(theta_history))     #一共走了多少步
    
```

```python
###2. 当将η设置为0.8时（原来为0.1）    出现了跳跃式的梯度下降，，步数为22，优化过程的时间缩短了，也最终找到了最优解
import numpy as np
import matplotlib.pyplot as plt

def J(theta):   #   损失函数
    return (theta -2.5)**2 -1
    
def dJ(theta):        #损失函数的导数
    return 2*(theta -2.5)

theta = 0.0   #初始点
theta_history = [theta]
eta = 0.8   #步长
epsilon = 1e-8  #精度问题或者eta的设置无法使得导数为0
while True:
    gradient = dJ(theta)    #求导数
    last_theta = theta      #先记录下上一个theta的值
    theta = theta - eta * gradient     #得到一个新的theta
    theta_history.append(theta)
    if(abs(J(theta) - J(last_theta)) < epsilon):
        break

        
plt.plot(plot_x,J(plot_x),color = 'r')
plt.plot(np.array(theta_history),J(np.array(theta_history)),color = 'b',marker = 'x')
plt.show()
print(len(theta_history))     #一共走了多少步
    
```
